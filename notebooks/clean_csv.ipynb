```python
import pandas as pd
# Loading csv as dataset
df = pd.read_csv("data/modified.csv")
df.head()
df['price'] = df['price'].fillna("Not Available")
unique_set = set()
for index, rows in df.iterrows():
    data = (rows["title"], rows["address"], rows["categoryName"], rows["city"], rows["location/lat"], rows["location/lng"], rows["price"], rows["reviewOrigin"], rows["reviewUrl"], rows["reviewerId"], rows["reviewsCount"], rows["street"], rows["totalScore"], rows["url"])
    unique_set.add(data)
# print(unique_set)
columns = ['title', 'address', 'categoryName', 'city', 'latitude', 'longitude', 'price', 'review_origin', 'review_url', 'rev_id', 'reviews_count', 'street', 'total_score', 'google_map_url']
unique_df = pd.DataFrame(list(unique_set), columns=columns)

# All things extracted except reviews which will be added as dictionary
unique_df
```
